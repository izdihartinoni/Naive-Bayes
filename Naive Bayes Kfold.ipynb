```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV 
from sklearn import datasets
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv("bank market.csv",sep=",")
data.head()
```

```python
data['y']=data['y'].map({'no':0,'yes':1})
y1=data['y']
x1=data.drop(columns='y')
x=x1.values
y=y1.values
```

```python
n=5
from sklearn.model_selection import StratifiedKFold
kf=StratifiedKFold(n_splits=n,random_state=None)
```

```python
from sklearn import datasets 
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB, ComplementNB
```

```python
cm1=[]
total1=[]
ac1=[]
se1=[]
sp1=[]
cm2=[]
total2=[]
ac2=[]
se2=[]
sp2=[]
cm3=[]
total3=[]
ac3=[]
se3=[]
sp3=[]
cm4=[]
total4=[]
ac4=[]
se4=[]
sp4=[]
```

```python
for train_index,test_index in kf.split(x,y):
    x_train, x_test =x[train_index],x[test_index]
    y_train, y_test =y[train_index],y[test_index]
    gnb = GaussianNB()
    brn = BernoulliNB()
    multi = MultinomialNB()
    comp = ComplementNB()
    y_pred = gnb.fit(x_train, y_train).predict(x_test)
    y1_pred=brn.fit(x_train,y_train).predict(x_test)
    y2_pred=multi.fit(x_train,y_train).predict(x_test)
    y3_pred=comp.fit(x_train,y_train).predict(x_test)
    cm1.append((confusion_matrix(y_test,y_pred)).astype(float))
    cm2.append((confusion_matrix(y_test,y1_pred)).astype(float))
    cm3.append((confusion_matrix(y_test,y2_pred)).astype(float))
    cm4.append((confusion_matrix(y_test,y3_pred)).astype(float))
for j in range (n):
    total1.append(sum(sum(cm1[j])))
    ac1.append((cm1[j][0,0]+cm1[j][1,1])/total1[j])
    se1.append(cm1[j][0,0]/(cm1[j][0,0]+cm1[j][0,1]))
    sp1.append(cm1[j][1,1]/(cm1[j][1,0]+cm1[j][1,1]))
for j in range (n):
    total2.append(sum(sum(cm2[j])))
    ac2.append((cm2[j][0,0]+cm2[j][1,1])/total2[j])
    se2.append(cm2[j][0,0]/(cm2[j][0,0]+cm2[j][0,1]))
    sp2.append(cm2[j][1,1]/(cm2[j][1,0]+cm2[j][1,1]))
for j in range (n):
    total3.append(sum(sum(cm3[j])))
    ac3.append((cm3[j][0,0]+cm3[j][1,1])/total3[j])
    se3.append(cm3[j][0,0]/(cm3[j][0,0]+cm3[j][0,1]))
    sp3.append(cm3[j][1,1]/(cm3[j][1,0]+cm3[j][1,1]))
for j in range (n):
    total4.append(sum(sum(cm4[j])))
    ac4.append((cm4[j][0,0]+cm4[j][1,1])/total4[j])
    se4.append(cm4[j][0,0]/(cm4[j][0,0]+cm4[j][0,1]))
    sp4.append(cm4[j][1,1]/(cm4[j][1,0]+cm4[j][1,1]))
akurasi1 = np.mean(ac1)
spesifisiti1 = np.mean(sp1)
sensitiviti1 = np.mean(se1)
akurasi2 = np.mean(ac2)
spesifisiti2 = np.mean(sp2)
sensitiviti2 = np.mean(se2)
akurasi3 = np.mean(ac3)
spesifisiti3 = np.mean(sp3)
sensitiviti3 = np.mean(se3)
akurasi4 = np.mean(ac4)
spesifisiti4 = np.mean(sp4)
sensitiviti4 = np.mean(se4)

```

```python
print("Gaussian Method")
print(akurasi1)
print(spesifisiti1)
print(sensitiviti1)
```

```python
print("Bernoulli Method")
print(akurasi2)
print(spesifisiti2)
print(sensitiviti2)
```

```python
print("Multinomial Method")
print(akurasi3)
print(spesifisiti3)
print(sensitiviti3)
```

```python
print("Complete Method")
print(akurasi4)
print(spesifisiti4)
print(sensitiviti4)
```
